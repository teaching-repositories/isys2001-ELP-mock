````python
# @title
distance_travelled =  100#@param {type:"integer"}
rental_days = 10 #@param {type:"integer"}

cost_per_km = 0.15 #@param {type:"number"}
cost_per_day = 20 #@param {type:"number"}

def calculate_cost(distance, days, cost_per_km, cost_per_day):
    """
    Calculates the total cost of the trip.

    The cost is calculated as the product of the distance and cost per kilometer
    plus the product of the days and cost per day, minus any applicable discount.

    Parameters:
    distance (int): The distance travelled in kilometers.
    days (int): The number of rental days.
    cost_per_km (float): The cost per kilometer.
    cost_per_day (float): The cost per day.

    Returns:
    int: The total cost of the trip.
    """
    discount = calculate_discount(distance, days)
    cost = distance * cost_per_km + days * cost_per_day - discount
    return round(cost)

def calculate_discount(distance, days):
    """
    Calculates the discount based on the distance travelled and the number of days.

    A discount of $100 is applied if the number of days is greater than 10 or if 
    the distance travelled is greater than 5000 kilometers.

    Parameters:
    distance (int): The distance travelled in kilometers.
    days (int): The number of rental days.

    Returns:
    int: The discount amount.
    """
    if days > 10 or distance > 5000:
        return 100
    return 0

def display_summary(distance, days, total_cost):
    """
    Displays a summary of the cost breakdown.

    Parameters:
    distance (int): The distance travelled in kilometers.
    days (int): The number of rental days.
    total_cost (int): The total cost of the trip.
    """
    discount = calculate_discount(distance, days)
    print("Distance:", distance, "kilometers")
    print("Days:", days, "days")
    if discount > 0:
        print("Discount: $", discount, sep="")
    print("Total: $", total_cost, sep="")

def main():
    """
    Main function to run the program.

    This function takes the input from the form, 
    calculates the total cost and displays a summary of the cost breakdown.
    """
    total_cost = calculate_cost(distance_travelled, rental_days, cost_per_km, cost_per_day)
    display_summary(distance_travelled, rental_days, total_cost)

main()
````
